```python
db = client['Supplements2']
collection1 = db["CompanyInformation"]
duplicate_docs = collection1.aggregate([
    { "$group": {
        "_id": { "DSLD ID": "$DSLD ID" },
        "uniqueIds": { "$addToSet": "$_id" },
        "count": { "$sum": 1 }
    }},
    { "$match": {
        "count": { "$gt": 1 }
    }},
    { "$project": {
        "duplicateIds": { "$slice": [ "$uniqueIds", 1, { "$size": "$uniqueIds" } ] }
    }}
])

# Delete the duplicate documents
for doc in duplicate_docs:
    collection1.delete_many({ "_id": { "$in": doc['duplicateIds'] } })
```

```python
#check duplicates
collection1 = db['CompanyInformation']

# Group documents by the 'DSLD ID' field and count the number of occurrences
pipeline = [
    {'$group': {'_id': '$DSLD ID', 'count': {'$sum': 1}}},
    {'$match': {'count': {'$gt': 1}}}
]
results = collection1.aggregate(pipeline, allowDiskUse=True)

# Print the duplicate DSLD IDs
for result in results:
    print(result['_id'])
```

## Create Index and define mapping

```python
# Define the index name and mapping
index_name = "vitamin_index"
mapping = {
    "mappings": {
        "properties": {
            "Product Name": {"type": "text"},
            "URL": {"type": "keyword"},
            "Company Name": {"type": "keyword"},
            "Country": {"type": "keyword"},
            "Ingredient": {"type": "text"},
            "DSLD ID": {"type": "text"},
            "DSLD Ingredient Categories": {"type": "keyword"},
            "Amount Per Serving": {"type": "float"},
            "Amount Per Serving Unit": {"type": "keyword"},
            "% Daily Value per Serving": {"type": "float"},
            "Daily Value Target Group": {"type": "keyword"},
            "Statement Type": {"type": "keyword"},
            "Statement": {"type": "text"},
            "Other Ingredients": {"type": "text"},
            "Brand Name": {"type": "keyword"},
            "Net Contents": {"type": "keyword"},
            "Serving Size": {"type": "keyword"},
            "Product Type [LanguaL]": {"type": "keyword"},
            "Supplement Form [LanguaL]": {"type": "keyword"},
            "Suggested Use": {"type": "text"}
        }
    }
}


# Delete the index if it already exists
if es.indices.exists(index_name):
    es.indices.delete(index_name)

# Create the Elasticsearch index and mapping
es.indices.create(index=index_name, body=mapping)
```

### Define a function to convert each filtered dataframe to an Elasticsearch document and index it

```python
def index_df(df, index_name):
    for i, row in df.iterrows():
        doc = row.to_dict()
        doc_id = f"{index_name}_{i}"
        es.index(index=index_name, id=doc_id, body=doc)
```

```python
from elasticsearch import Elasticsearch
from flask import Flask, render_template, request
import numpy as np
import pandas as pd

# Initialize Flask app
app = Flask("vitamin")


# Get a list of unique values from the 'Product Name' column of each dataframe
product_names = df1.iloc[:,2].tolist()

# Initialize Elasticsearch client
es= Elasticsearch(['http://localhost:9200'])

# Define the columns to display in each table
cols1 = ['DSLD ID','Product Name','URL','Company Name','Country']
cols2 = ['DSLD ID','Ingredient', 'DSLD Ingredient Categories', 'Amount Per Serving', 'Amount Per Serving Unit', '% Daily Value per Serving', 'Daily Value Target Group']
cols3 = ['DSLD ID','Statement Type', 'Statement']
cols4 = ['DSLD ID','Other Ingredients']
cols5 = ['DSLD ID','Brand Name', 'Net Contents','Serving Size','Product Type [LanguaL]','Supplement Form [LanguaL]','Suggested Use']


# Set index name and create index with mapping

common_col = "Product Name"

@app.route('/', methods=['GET', 'POST'])
def index():
    if request.method == 'POST':
        # Get the search input from the dropdown
          # Get the search input from the dropdown
        search_input = request.form['product_names']

        # Search for documents in Elasticsearch that match the search input in the 'Product Name' field
        query = {
            "query": {
                "match": {
                    "Product Name": search_input
                }
            }
        }
        res = es.search(index=index_name, body=query, size = 20)

        # Get a list of the matching product names
        hits = res['hits']['hits']
        matching_product_names = [hit['_source'][common_col] for hit in hits]

        # Filter each dataframe based on the matching product names
        df1_filtered = df1[df1[common_col].isin(matching_product_names)][cols1]
        df2_filtered = df2[df2[common_col].isin(matching_product_names)][cols2]
        df3_filtered = df3[df3[common_col].isin(matching_product_names)][cols3]
        df4_filtered = df4[df4[common_col].isin(matching_product_names)][cols4]
        df5_filtered = df5[df5[common_col].isin(matching_product_names)][cols5]

        # Convert each filtered dataframe to an HTML table
        table1 = df1_filtered.to_html(classes='data')
        table2 = df2_filtered.to_html(classes='data')
        table3 = df3_filtered.to_html(classes='data')
        table4 = df4_filtered.to_html(classes='data')
        table5 = df5_filtered.to_html(classes='data')

        # Render the result template with the HTML tables
        return render_template('result.html', table1=table1, table2=table2, table3=table3, table4=table4, table5=table5)

 # Render the index template on initial GET request
    return render_template('index.html',  product_names=product_names)
```
